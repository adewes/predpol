```python
%matplotlib inline
from matplotlib.pylab import *
import numpy as np
from matplotlib import gridspec

import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'
mpl.rcParams['font.size'] = 13
```

```python
def sigmoid(t):
    return 1.0/(1.0+np.exp(-t))

gs = gridspec.GridSpec(2,1, width_ratios=[2, 1]) 
x = np.arange(-10,10,0.01)
figure(figsize=(12,8))
subplot(gs[0])
xlabel("t")
ylabel("probability")
plot(x, sigmoid(x))
axhline(0.5, ls='--',color='grey')
axvline(0, ls='--',color='grey')
subplot(gs[1])
y = np.arange(0,30, 0.1)
l = len(y)//2
plot(-12.4+y[:l]*0.9+0.2*y[:l]*np.random.randn(l),y[:l],marker='o',ls='')
plot(-12.4+y[l:]*0.9+0.2*y[l:]*np.random.randn(l),y[l:],marker='o',ls='', color='green')
plot(-12.4+y*0.9,y,color='red')
axvline(0, ls='--',color='grey')
axhline(14.2, ls='--',color='grey')
xlabel("hidden variable t")
ylabel("observed variable x")
legend(["input data", "best linear fit"],loc ='upper left')
xlim(-10,10)
```
